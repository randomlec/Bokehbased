```python
from bokeh.models.markers import Triangle
from bokeh.models.sources import ColumnDataSource
from bokeh.layouts import row
import numpy as np
from ipywidgets import interact
from bokeh.plotting import figure, output_file, show
from bokeh.io import show, curdoc, push_notebook, output_notebook
output_notebook()

x = np.linspace(0,10,200)
y = x

p = figure(title="Savings per capita", plot_height=300, plot_width=400, y_range=(0,5))
r = p.line(x, y, color="#2222aa", line_width=3, legend='r')
t = p.line(x,y,color="#ff0000", line_width=3, legend='t')
o = p.line(x,y,color="#009933", line_width=3,legend='o')
tri = p.scatter(x,y, color='orange',line_width=5,marker='triangle',size=20, legend='triangle')
p.legend.location = "top_left"
p.legend.border_line_alpha = 0
p.xaxis.axis_label = "Lot Number"
p.yaxis.axis_label = "Bin Count"

p2 = figure(title="Savings per capita", plot_height=300, plot_width=400, y_range=(0,5))
r2 = p2.line(x, y, color="olive", line_width=3,legend='r2')
t2 = p2.line(x,y,color="purple", line_width=3,legend='t2')
o2 = p2.line(x,y,color="cyan", line_width=3,legend='o2')
p2.legend.location = "top_left"
p2.legend.border_line_alpha = 0
p2.xaxis.axis_label = "Lot Number"
p2.yaxis.axis_label = "Bin Count"

def update(s=0.5, z=1, =0.3, n=0.2, =0.2):
    r.data_source.data['y'] = s*z*x**
    t.data_source.data['y'] = z*x**
    o.data_source.data['y'] = (+n)*x
    
    r2.data_source.data['y'] = s*z*x**
    t2.data_source.data['y'] = z*x**
    o2.data_source.data['y'] = (+n)*x
    
    kss = ((s*z)/(n+))**(1/(1-))
    sss = s*z*kss**
    tri.data_source.data['x'] = [kss]
    tri.data_source.data['y'] = [sss]
    push_notebook()

layout = row(p,p2)
show(layout, notebook_handle=True)
interact(update, s=(0.01,0.8), z=(1.2,2), =(0.2, 0.8, 0.1), n=(0.01,0.5,0.05), =(0.01,0.5,0.05))
```
